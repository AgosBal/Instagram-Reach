```python
#Import necessary libraries

import pandas as pd 
import matplotlib.pyplot as plt #Library to create graphics and view data in Python
from datetime import datetime, timedelta   #To work with dates and hours   
from pytrends.request import TrendReq #Library that allows to access to Google Trends

#Create an instance of TrendReq, and assign it to the variable "Trending_topics"
Trending_topics = TrendReq(hl = 'en-US', tz=360, geo='') 
#First parameter hl is for language and location
#Second parameter tz is for time zone in minutes. 360 == 6 hs
#Third parameter geo is to search globally
```

```python
#Create a dataframe for top countries with keyword "time to publish"

keyword = ["Time to publish"]
Trending_topics.build_payload(keyword, cat=0, timeframe='today 12-m') #build_payload is a method used to configure and build a search trend query with the following parameters:
                                                                      #keyword: keyword or phrase of the query, which  from the "keyword" list.
                                                                      #cat=0: used to specify the category to which the search belongs. It is set to 0, indicating that the search is not being restricted to a specific category
                                                                      #timeframe='today 12-m': define the time period for which to obtain the search trends. 'today 12-m' means it will get trend data for the last year as of today.                                                                  
                        
```

```python
#Obtain search trend data related to the keyword "Time to publish" over the last year.

Trending_topics.build_payload(kw_list=["Time to publish"], cat=0, timeframe='today 12-m') 
data = Trending_topics.interest_over_time() #Queries Google Trends using the previous configuration and stores the data in the DataFrame "data"
data = data.sort_index()
```

```python
#Graphic
plt.figure(figsize=(12, 6))
plt.plot(data.index, data["Time to publish"], marker='o', linestyle='-', color='b')
plt.title("Search trends for last year: Time to publish")
plt.xlabel("Date")
plt.ylabel("Interest")
plt.grid(True)

#Show the graphic
plt.tight_layout()
plt.show()
```
